```python
import pandas as pd
import numpy as np
from datetime import timedelta



def estimar_proximo_pago(grupo_usuario_giro):
    """
    Estima la próxima fecha de pago y monto para un grupo de transacciones
    (de un usuario y un giro específicos).
    """
    df = grupo_usuario_giro.sort_values('fecha').copy()

    if len(df) < 2:
        return None, None, None # No se puede estimar con menos de 2 transacciones

    # Calcular diferencias en días
    df['diferencia_dias'] = df['fecha'].diff().dt.days

    # Filtrar diferencias razonables (ej. 15 a 45 días) y calcular mediana
    diferencias_filtradas = df['diferencia_dias'][(df['diferencia_dias'] >= 15) & (df['diferencia_dias'] <= 45)]

    if not diferencias_filtradas.empty:
        periodicidad = int(diferencias_filtradas.median())
    else:
        # Si no hay diferencias 'mensuales', intentamos con la media de todas > 0
        otras_diferencias = df['diferencia_dias'][df['diferencia_dias'] > 0]
        if not otras_diferencias.empty:
           periodicidad = int(otras_diferencias.mean())
           # Acotamos a un máximo (ej 60 días) para evitar valores raros
           periodicidad = min(periodicidad, 60)
        else:
           periodicidad = 30 # Último recurso: 30 días

    # Asegurarnos que la periodicidad sea al menos 1 día
    periodicidad = max(periodicidad, 1)

    ultima_transaccion = df.iloc[-1]
    fecha_ultimo_pago = ultima_transaccion['fecha']
    monto_estimado = ultima_transaccion['monto']
    comercio = ultima_transaccion['comercio']

    fecha_proximo_pago = fecha_ultimo_pago + timedelta(days=periodicidad)

    return fecha_proximo_pago, monto_estimado, comercio

def enviar_alertas_recurrentes(df_transacciones, hoy_simulado, giros_interes, dias_alerta=3):
    """
    Analiza 'df_transacciones' y envía alertas/recordatorios para pagos
    próximos basados en la fecha simulada 'hoy'.
    """
    hoy = pd.to_datetime(hoy_simulado)
    print(f"\n--- Iniciando revisión de alertas para el día: {hoy.strftime('%Y-%m-%d')} ---")

    # Asegurarse que 'fecha' sea datetime (por si acaso)
    df_transacciones['fecha'] = pd.to_datetime(df_transacciones['fecha'])

    # Filtrar solo los giros que nos interesan
    df_interes = df_transacciones[df_transacciones['giro_DA'].isin(giros_interes)].copy()

    if df_interes.empty:
        print("No se encontraron transacciones para los giros de interés especificados.")
        return

    # Obtener el último saldo conocido por usuario
    df_sorted = df_transacciones.sort_values('fecha')
    ultimo_saldo_map = df_sorted.groupby('id')['Saldo'].last().to_dict()

    # Agrupar por usuario y giro de interés
    grupos = df_interes.groupby(['id', 'giro_DA'])

    alertas_enviadas = 0

    for (user_id, giro), grupo_transacciones in grupos:
        # Aquí, 'grupo_transacciones' es el DataFrame con las transacciones
        # solo para 'user_id' y 'giro'. Lo pasamos a la función:
        proxima_fecha, monto_estimado, comercio = estimar_proximo_pago(grupo_transacciones)

        if proxima_fecha and monto_estimado is not None:
            dias_faltantes = (proxima_fecha - hoy).days

            # Verificar si el pago es en los próximos 'dias_alerta' (incluyendo hoy)
            if 1 <= dias_faltantes <= dias_alerta:
                saldo_usuario = ultimo_saldo_map.get(user_id, 0)
                alertas_enviadas += 1

                if saldo_usuario < monto_estimado:
                    print(f"[ALERTA 🚨] Usuario: ...{user_id[-6:]}, Giro: {giro}")
                    print(f"  > Tu cargo por '{comercio}' (${monto_estimado:.2f}) se cobrará en {dias_faltantes} día(s) pero tu saldo (${saldo_usuario:.2f}) es insuficiente 😰.")
                    print(f"  > No lo dejes pasar, entra a Hey Banco App para ver los cajeros más cercanos 📍\n")
                else:
                    print(f"[RECORDATORIO 💡] Usuario: ...{user_id[-6:]}, Giro: {giro}")
                    print(f"  > Tu cargo por '{comercio}' (${monto_estimado:.2f}) se cobrará en {dias_faltantes} día(s) ¡no lo olvides! 🤗\n")

    print(f"--- Revisión completada. Se encontraron {alertas_enviadas} posibles alertas/recordatorios. ---")


# --- Ejecución del Proceso ---

# 1. Lista de giros que consideras recurrentes
giros_recurrentes = [
    'Telefonía móvil', 'Internet / TV de paga', 'Streaming de video',
    'Streaming de música', 'Suscripción digital', 'Suscripción de comida',
    'Servicios públicos'
]

# 2. Definir una fecha "hoy" para simular (dentro del rango de tus datos)
fecha_simulada = '2022-12-10' # Cambia esta fecha para probar diferentes escenarios

# 3. Ejecutar la función de alertas (usa tu 'trans_df' real aquí)
#    Nosotros usamos el 'trans_df' creado en la simulación.
enviar_alertas_recurrentes(trans_df, fecha_simulada, giros_recurrentes)
```

Este código tiene como objetivo principal simular un sistema de notificaciones proactivo para los clientes de un banco (en este caso, Hey Banco). Su función es identificar los pagos que parecen ser recurrentes (como suscripciones o servicios) y enviar alertas o recordatorios a los usuarios cuando la fecha de cobro estimada está muy próxima (en 3 días o menos). La notificación varía dependiendo de si el saldo del cliente es suficiente o no para cubrir el gasto estimado.

¿Cómo lo logra? Se divide en dos partes principales: una función de apoyo para estimar pagos y una función principal para orquestar las alertas.

1. estimar_proximo_pago(grupo_usuario_giro): La Lógica Predictiva

Esta función es el cerebro detrás de la predicción. Recibe un conjunto de transacciones que pertenecen a un solo usuario y a un solo tipo de gasto recurrente (por ejemplo, todas las transacciones de "Juan Pérez" en "Streaming de video").

Requisito Mínimo: Si no hay al menos dos transacciones, es imposible calcular un patrón, por lo que no hace nada.
Cálculo de Días: Ordena las transacciones por fecha y calcula cuántos días pasaron entre cada una.
Búsqueda de Periodicidad: Intenta encontrar un patrón "típico" de pago. Busca diferencias de días que sean lógicas para suscripciones (entre 15 y 45 días) y calcula la mediana (el valor central). La mediana es útil para evitar que pagos muy raros o errores afecten la estimación. Si no encuentra un patrón claro "mensual", intenta con la media de todas las diferencias (limitada a 60 días) y, como último recurso, asume 30 días.
Estimación Final: Una vez que tiene una periodicidad (por ejemplo, 30 días), la suma a la fecha de la última transacción registrada. Esto le da una fecha estimada para el próximo cobro. También recupera el monto y el comercio de esa última transacción para usarlos en el mensaje.
Resultado: Devuelve la fecha estimada, el monto estimado y el comercio.

2. enviar_alertas_recurrentes(df_transacciones, hoy_simulado, giros_interes, dias_alerta=3): El Orquestador de Notificaciones

Esta es la función principal que organiza todo el proceso.

Definición del "Hoy": Establece una fecha (hoy_simulado) que actúa como el "día actual" para realizar las comprobaciones. Esto es crucial porque los datos son históricos.
Filtrado Inicial: Se concentra solo en las transacciones que pertenecen a los giros que nos interesan (giros_interes, como 'Telefonía móvil', 'Streaming', etc.).
Obtención de Saldos: Para cada usuario, busca cuál fue su último saldo registrado en toda su historia de transacciones. Este será el saldo que usará para comparar.
