```python
df['fecha'] = pd.to_datetime(df['fecha'])
fecha_ref = df['fecha'].max()

# Calcular fechas clave
df_sorted = df.sort_values(['id', 'giro_DA', 'fecha'])

# Días entre compras
df_sorted['dias_entre_compras'] = df_sorted.groupby(['id', 'giro_DA'])['fecha'].diff().dt.days

# Tabla resumen por cliente-giro
df_giro = df_sorted.groupby(['id', 'giro_DA']).agg(
    n_trans=('fecha', 'count'),
    monto_total=('monto', 'sum'),
    primera_fecha=('fecha', 'min'),
    ultima_fecha=('fecha', 'max'),
    promedio_entre_compras=('dias_entre_compras', 'mean')
).reset_index()

df_giro['dias_desde_primera'] = (fecha_ref - df_giro['primera_fecha']).dt.days
df_giro['dias_ultima'] = (fecha_ref - df_giro['ultima_fecha']).dt.days
df_giro
```

## Indice de "Favoritismo"

```python
df_giro['frecuencia_pct'] = df_giro.groupby('id')['n_trans'].transform(lambda x: x / x.max())
df_giro['monto_pct'] = df_giro.groupby('id')['monto_total'].transform(lambda x: x / x.max())

# Índice ponderado
df_giro['favoritismo'] = (0.8 * df_giro['frecuencia_pct']) + (0.2 * df_giro['monto_pct'])
df_giro
```

```python
# Giros que NO deben ser considerados para 'tus_favoritos'
giros_excluidos = [
    'Servicios públicos',
    'Telefonía móvil',
    'Gestión de pagos móviles',
    'Fintech / BNPL',
    'Otros',
    'Suscripción digital',
    'Streaming de video'
]
df_giro_filtrado = df_giro[~df_giro['giro_DA'].isin(giros_excluidos)].copy()
df_giro_filtrado['ranking_favoritismo'] = df_giro_filtrado.groupby('id')['favoritismo'].rank(method='dense', ascending=False)
favoritos_top2 = df_giro_filtrado[df_giro_filtrado['ranking_favoritismo'] <= 2].copy()
favoritos_top2['tipo_gancho'] = 'tus_favoritos'

```

```python
# Asegúrate de tener esta columna de fecha en tu df original
df['fecha'] = pd.to_datetime(df['fecha'])
fecha_ref = pd.to_datetime('2022-12-01')  # o ajusta a tu fecha objetivo

# Año-mes para agrupar
df['año_mes'] = df['fecha'].dt.to_period('M')
df['mes_diff'] = ((fecha_ref.to_period('M') - df['año_mes']).apply(lambda x: x.n))

# Conteo mensual por cliente-giro
conteo_mensual = df.groupby(['id', 'giro_DA', 'año_mes']).size().reset_index(name='transacciones')

# Promedio histórico (meses 3 a 5 atrás)
historico = conteo_mensual[conteo_mensual['año_mes'].apply(lambda x: 3 <= (fecha_ref.to_period('M') - x).n <= 5)]
prom_hist = historico.groupby(['id', 'giro_DA'])['transacciones'].mean().reset_index(name='trans_promedio')

# Promedio últimos 2 meses
ultimos2 = conteo_mensual[conteo_mensual['año_mes'].apply(lambda x: 1 <= (fecha_ref.to_period('M') - x).n <= 2)]
prom_ult2 = ultimos2.groupby(['id', 'giro_DA'])['transacciones'].mean().reset_index(name='trans_prom_ult2m')

# Total últimos 2 meses
total_ult2 = ultimos2.groupby(['id', 'giro_DA'])['transacciones'].sum().reset_index(name='trans_total_ult2m')

# Total último mes (mes 1 atrás)
ultimo_mes = conteo_mensual[conteo_mensual['año_mes'].apply(lambda x: (fecha_ref.to_period('M') - x).n == 1)]
total_ultimo = ultimo_mes.groupby(['id', 'giro_DA'])['transacciones'].sum().reset_index(name='trans_total_ult_mes')

# Agregar al df_giro
df_giro = df_giro.merge(prom_hist, on=['id', 'giro_DA'], how='left') \
                 .merge(prom_ult2, on=['id', 'giro_DA'], how='left') \
                 .merge(total_ult2, on=['id', 'giro_DA'], how='left') \
                 .merge(total_ultimo, on=['id', 'giro_DA'], how='left')
df_giro
```

```python
# crece_vs_hist: si promedio histórico es NaN y total no lo es, entonces True; si ambos son NaN, False
df_giro['crece_vs_hist'] = (
    (df_giro['trans_total_ult2m'] > df_giro['trans_promedio']) |
    ((df_giro['trans_promedio'].isna()) & (df_giro['trans_total_ult2m'].notna()))
)

# crece_vs_reciente: si promedio reciente es NaN y total no lo es, entonces True; si ambos son NaN, False
df_giro['crece_vs_reciente'] = (
    (df_giro['trans_total_ult_mes'] > df_giro['trans_prom_ult2m']) |
    ((df_giro['trans_prom_ult2m'].isna()) & (df_giro['trans_total_ult_mes'].notna()))
)
```

```python
# Rankear todos los giros por cliente según favoritismo
df_giro['ranking_favoritismo'] = df_giro.groupby('id')['favoritismo'].rank(method='dense', ascending=False)

# Filtrar lógica de "vas que vuelas"
vas_que_vuelas = df_giro[
    (~df_giro['giro_DA'].isin(giros_excluidos)) &
    (df_giro['ranking_favoritismo'].between(3, 5)) &
    (df_giro['crece_vs_reciente'] == True)
]

# Excluir giros ya clasificados como favoritos
vas_que_vuelas = vas_que_vuelas.merge(
    favoritos_top2[['id', 'giro_DA']],
    on=['id', 'giro_DA'],
    how='left',
    indicator=True
)
vas_que_vuelas = vas_que_vuelas[vas_que_vuelas['_merge'] == 'left_only'].drop(columns=['_merge'])

# Etiquetar tipo de gancho
vas_que_vuelas['tipo_gancho'] = 'vas_que_vuelas'
vas_que_vuelas.id.nunique()
# 374 clientes con crecimiento en ciertos giros.
```

```python
# Máximo intervalo entre compras por cliente-giro (solo entre la 2ª y la penúltima)
max_intervalo = df_sorted.groupby(['id', 'giro_DA'])['dias_entre_compras'].max().reset_index(name='max_dias_entre_compras')

# Unir al df_giro
df_giro = df_giro.merge(max_intervalo, on=['id', 'giro_DA'], how='left')
```

```python
# Unir favoritos y vas que vuelas
bloqueados = pd.concat([
    favoritos_top2[['id', 'giro_DA']],
    vas_que_vuelas[['id', 'giro_DA']]
], ignore_index=True).drop_duplicates()

# Excluir bloqueados y giros excluidos
df_filtrado = df_giro.merge(
    bloqueados,
    on=['id', 'giro_DA'],
    how='left',
    indicator=True
)
df_filtrado = df_filtrado[
    (df_filtrado['_merge'] == 'left_only') &
    (~df_filtrado['giro_DA'].isin(giros_excluidos))
].drop(columns=['_merge'])
df_filtrado['ranking_favoritismo'] = df_filtrado.groupby('id')['favoritismo'].rank(method='dense', ascending=False)
y_si_regresas_candidatos = df_filtrado[df_filtrado['ranking_favoritismo'] <= 2]
y_si_regresas = y_si_regresas_candidatos[
    (y_si_regresas_candidatos['dias_ultima'] >= y_si_regresas_candidatos['max_dias_entre_compras'] - 5)
]
y_si_regresas['tipo_gancho'] = 'y_si_regresas'
#293 clientes con algún y si regresas activo
y_si_regresas
```

```python
giros_excluidos_yamerito = [
    'Servicios públicos',
    'Telefonía móvil',
    'Gestión de pagos móviles',
    'Fintech / BNPL',
    'Suscripción digital',
    'Streaming de video',
    'Streaming de música',
    'Otros'
]
# Base: df_giro ya tiene dias_ultima, promedio_entre_compras y n_trans
ya_merito = df_giro[
    (df_giro['n_trans'] >= 3) &
    (~df_giro['giro_DA'].isin(giros_excluidos_yamerito)) &
    (df_giro['promedio_entre_compras'].notna()) &
    (abs(df_giro['dias_ultima'] - df_giro['promedio_entre_compras']) <= 2)
]

# Excluir giros ya etiquetados en otros ganchos
bloqueados = pd.concat([
    favoritos_top2[['id', 'giro_DA']],
    vas_que_vuelas[['id', 'giro_DA']],
    y_si_regresas[['id', 'giro_DA']]
], ignore_index=True).drop_duplicates()

ya_merito = ya_merito.merge(
    bloqueados,
    on=['id', 'giro_DA'],
    how='left',
    indicator=True
)
ya_merito = ya_merito[ya_merito['_merge'] == 'left_only'].drop(columns=['_merge'])

# Etiquetar
ya_merito['tipo_gancho'] = 'ya_merito'
ya_merito
```
