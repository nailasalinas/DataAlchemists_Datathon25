```python
# Lista de giros considerados suscripciones
giros_suscripcion = [
    'Streaming de música',
    'Streaming de video',
    'Suscripción de comida',
    'Telefonía móvil',
    'Internet / TV de paga',
    'Suscripción digital',
    'Plataforma de diseño'
]

# Filtrar solo transacciones de suscripción
df_suscripciones = df[df['giro_DA'].isin(giros_suscripcion)]

# Obtener número de clientes únicos con al menos una transacción en suscripciones
num_clientes_con_suscripcion = df_suscripciones['id'].nunique()

print(f"Clientes con al menos una transacción en suscripciones: {num_clientes_con_suscripcion}")
```

```python
# Lista de giros relacionados con suscripciones
giros_suscripcion = [
    'Streaming de música',
    'Streaming de video',
    'Suscripción de comida',
    'Telefonía móvil',
    'Internet / TV de paga',
    'Suscripción digital',
    'Plataforma de diseño'
]

# Gasto total (toda la base)
gasto_total = df['monto'].sum()

# Gasto total solo en giros de suscripción
gasto_suscripciones = df[df['giro_DA'].isin(giros_suscripcion)]['monto'].sum()

# Porcentaje del gasto en suscripciones respecto al total
porcentaje_suscripciones = (gasto_suscripciones / gasto_total) * 100

# Imprimir resultado
print(f"Porcentaje del monto en giros de suscripción respecto al total: {porcentaje_suscripciones:.2f}%")
```
